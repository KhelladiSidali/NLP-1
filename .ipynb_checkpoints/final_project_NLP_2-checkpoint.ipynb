```python
def compute_unigram_and_bigram_probs(tags):
    # Calcul des fréquences des unigrammes
    unigram_counts = Counter(tags)
    total_unigrams = sum(unigram_counts.values())
    
    # Calcul des probabilités des unigrammes
    unigram_probs = {tag: count / total_unigrams for tag, count in unigram_counts.items()}
    
    # Calcul des fréquences des bigrammes
    bigram_counts = Counter(zip(tags, islice(tags, 1, None)))
    total_bigrams = sum(bigram_counts.values())
    
    # Calcul des probabilités des bigrammes
    bigram_probs = {bigram: count / total_bigrams for bigram, count in bigram_counts.items()}

    return unigram_probs, bigram_probs
ner_tag_unigram_probs, ner_tag_bigram_probs = compute_unigram_and_bigram_probs(ner_tag_list)
```

```python
# fonction qui retourn un objet qui contient les tokens avec leurs NER

def annotate_sentences(sentences):
    annotated_data = []
    for sentence in sentences:
        doc = nlp(sentence)
        annotated_sentence = [(token.text, token.ent_iob_ + '-' + token.ent_type_ if token.ent_type_ else 'O') for token in doc]
        annotated_data.append(annotated_sentence)
    return annotated_data
annotated_data = annotate_sentences(sentences)
annotated_data
```

```python
# mettre tout les NER dans une liste (avec repitition)
# pour calculer unigram & bigram

def extract_ner_tags(annotated_data):
    ner_tags = []
    for sentence in annotated_data:
        for token, tag in sentence:
            if tag != 'O':
                ner_tags.append(tag)
    return ner_tags
extract_ner_tags(annotated_data)
```

```python
def compute_ner_transition_matrices(tags):
    unigram_counts = Counter(tags)
    bigram_counts = Counter(zip(tags, tags[1:]))

    total_unigrams = sum(unigram_counts.values())
    total_bigrams = sum(bigram_counts.values())

    unigram_probs = {tag: count / total_unigrams for tag, count in unigram_counts.items()}
    bigram_probs = {bigram: count / total_bigrams for bigram, count in bigram_counts.items()}

    return unigram_probs, bigram_probs
```
